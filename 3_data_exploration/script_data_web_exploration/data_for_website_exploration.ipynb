```python
import pandas as pd
```

```python
#Load csv files
df1=pd.read_csv("2nd dataset.csv")
# Load Excel file
df2 = pd.read_excel ("data_other_site.xlsx")
df3 = pd.read_excel("1st dataset_draft...xlsx")
df4 = pd.read_excel ("data_tourist_site.xlsx")
```

```python
df2.columns = df2.iloc[1]
df2 = df2.iloc[1:].reset_index(drop =True)
df2
```

```python
df_all = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
df_all.columns
```

```python
titles = [
    "Aéroport", "Hôtel", "Plage", "Grotte", "Canyon",
       "Site Historique", "Site Culturel"
]
```

```python
columns_to_search = ["Category", "Name", "Type/Field", "Destination","Type"]

for title in titles:
    # Filter: keep rows where ANY of the columns matches the title
    filtered = df_all[df_all[columns_to_search].isin([title]).any(axis=1)]
    
    # Save output
    output_name = f"{title.replace(' ', '_')}.csv"
    filtered.to_csv(output_name, index=False)
    
    print(f"Saved → {output_name} ({len(filtered)} rows)")
```
