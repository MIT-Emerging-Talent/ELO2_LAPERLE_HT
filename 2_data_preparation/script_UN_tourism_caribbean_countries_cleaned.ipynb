```python
import pandas as pd
```

```python
# Load the csv file cleaning 
df= pd.read_csv("UN_tourism_caribbean_countries_raw.csv")
df.head()
```

```python
df.columns
```

```python
# drop empty and irrelevant columns to prepare the dataset for analysis 
df_caribbean = df.drop(columns=["indicator_code", "indicator_previous_code", "indicator_previous_code","reporter_area_code",
                                "partner_area_code", "flag", "flag_label","notes","drop"])
```

```python
# rename the visitors types in indicator_label for more clarity in simplicity 
df_caribbean["indicator_label"] = df_caribbean["indicator_label"].replace({
    "inbound - trips - total - cruise ships - same-day visitors (excursionists)": "excursionists",
    "inbound - trips - total - total - overnight visitors (tourists)": "tourists",
    "inbound - trips - total - total - same-day visitors (excursionists)": "excursionists",
    "inbound - trips - total - total - visitors": "total_visitors"
})
```

```python
df_caribbean.head()
```

```python
#see the remaining columns 
df_caribbean.columns
```

```python
# rename the remaining columns for clarity and simplicity
df_caribbean = df_caribbean.rename(columns={
    "indicator_label": "type_of_visitors",
    "reporter_area_label": "country_receiving",
    "partner_area_label": "where_tourist_from",
    "value": "number_of_tourist"
})
```

```python
df_caribbean.sample(10)
```

```python
# save the clean dataset in a csv file
df_caribbean.to_csv("UN_tourism_caribbean_countries_cleaned.csv",index=False)
```
