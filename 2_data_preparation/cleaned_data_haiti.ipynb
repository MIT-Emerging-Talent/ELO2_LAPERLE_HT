```python
import pandas as pd
```

```python
#------------------------------------------------------------------------------
# Step 1
#------------------------------------------------------------------------------
# Load the UN_tourism_inbound_Arrivals_haiti for cleaning
df= pd.read_csv("UN_tourism_inbound_Arrivals_haiti.csv")
df.head()
```

```python
# Dropping empty and irrelevant columns to clean the dataset for analysis  
df_arrivals_haiti = df.drop(columns=["indicator_code", "indicator_previous_code", "indicator_previous_code","reporter_area_code",
                                "partner_area_code", "flag", "flag_label","notes","drop"])
```

```python
# Rename indicator label categories to standardize types of visitors 
df_arrivals_haiti["indicator_label"] = df_arrivals_haiti["indicator_label"].replace({
    "inbound - trips - total - cruise ships - same-day visitors (excursionists)": "excursionists",
    "inbound - trips - total - total - overnight visitors (tourists)": "tourists",
    "inbound - trips - total - total - same-day visitors (excursionists)": "excursionists",
    "inbound - trips - total - total - visitors": "visitors_total"
})
```

```python
# rename columns for more clarity
df_arrivals_haiti = df_arrivals_haiti.rename(columns={
    "indicator_label": "type_of_visitors",
    "reporter_area_label": "country_receiving",
    "partner_area_label": "where_tourist_from",
    "value": "number_of_tourist",
    "unit" : "unit_thousand_trips"
})
```

```python
# visualize the first five(5) lines of df_arrivals_haiti dataset
df_arrivals_haiti.head()
```

```python
# see the shape of the dataset 
df_arrivals_haiti.shape
```

```python
# save the cleaned dataset in a csv file
df_arrivals_haiti.to_csv("UN_tourism_inbound_Arrivals_haiti_cleaned.csv",index=False)
```
